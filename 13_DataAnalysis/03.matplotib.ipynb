```python
import numpy as np

x = np.arange(0,12,1)
y = np.cos(x) # cos, sin, tan
y2 = np.sin(x)

plt.plot(x,y2, label='sin')
plt.plot(x,y, label='cos')
plt.legend(loc=(1,1))

plt.title('cos graph')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid()

plt.xlim(0, np.pi)
plt.ylim(-1, 1)
```
